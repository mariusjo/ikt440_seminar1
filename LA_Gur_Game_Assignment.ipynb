Make 20 gamesets. Increase nr of states of each gameset. Printout for each gameset is the mean value of nr of yes votes through all games in set.

```python
ggs = [GurGame(nr_of_states= 1 + v , nr_of_rounds=5000, nr_of_games=50) for v in range(20)]
print("all done!")
```
